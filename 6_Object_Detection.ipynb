```python
# PS 1:
# Apply a pretrained object detection model to identify objects in images.

import kagglehub
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os

# Download dataset
path = kagglehub.dataset_download("deeppratap/traffic-sign-detection-yolov8")

print("Path to dataset files:", path)

# Load pretrained YOLO model
model = YOLO("yolov8n.pt")

# Example image path
image_path = os.path.join(path, "train/images")

# Get first image
img_name = os.listdir(image_path)[0]
img_file = os.path.join(image_path, img_name)

# Run detection
results = model(img_file, imgsz=320)

# Plot output
result_img = results[0].plot()

plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
plt.title("YOLO Detection")
plt.show()
```

```python
# PS 2:
# Analyse confidence scores and detection accuracy for multiple objects.

import kagglehub
from ultralytics import YOLO
import os

# Download dataset
path = kagglehub.dataset_download("deeppratap/traffic-sign-detection-yolov8")

print("Path to dataset files:", path)

# Load YOLO model
model = YOLO("yolov8n.pt")

image_path = os.path.join(path, "train/images")

img_name = os.listdir(image_path)[0]
img_file = os.path.join(image_path, img_name)

results = model(img_file, imgsz=320)

boxes = results[0].boxes

print("Number of detections:", len(boxes))

for box in boxes:
    print("Confidence Score:", float(box.conf))
```

```python
# PS 3:
# Identify cases where detection fails and explain possible reasons.

import kagglehub
from ultralytics import YOLO
import os

# Download dataset
path = kagglehub.dataset_download("deeppratap/traffic-sign-detection-yolov8")

print("Path to dataset files:", path)

model = YOLO("yolov8n.pt")

image_path = os.path.join(path, "train/images")

img_name = os.listdir(image_path)[0]
img_file = os.path.join(image_path, img_name)

results = model(img_file, imgsz=320)

boxes = results[0].boxes

print("Objects Detected:", len(boxes))

print("Detection may fail because of:")
print("- Blur")
print("- Low lighting")
print("- Small objects")
print("- Occlusion")
```

```python
# PS 4:
# Evaluate how image quality affects detection performance.

import kagglehub
from ultralytics import YOLO
import cv2
import os

# Download dataset
path = kagglehub.dataset_download("deeppratap/traffic-sign-detection-yolov8")

print("Path to dataset files:", path)

model = YOLO("yolov8n.pt")

image_path = os.path.join(path, "train/images")

img_name = os.listdir(image_path)[0]
img_file = os.path.join(image_path, img_name)

# Original image
img = cv2.imread(img_file)

# Blur image
blur_img = cv2.GaussianBlur(img, (15,15), 0)

# Save blurred image
cv2.imwrite("blur.jpg", blur_img)

# Detect on original
result1 = model(img_file, imgsz=320)

# Detect on blurred
result2 = model("blur.jpg", imgsz=320)

print("Clear Image Detections:", len(result1[0].boxes))
print("Blur Image Detections:", len(result2[0].boxes))

print("Poor image quality reduces detection performance.")
```
