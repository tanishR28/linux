```python
# NumPy -> Numerical operations
import numpy as np

# Matplotlib -> Graph plotting
import matplotlib.pyplot as plt

# make_classification -> Generates synthetic dataset
from sklearn.datasets import make_classification

# LDA -> Dimensionality Reduction Technique
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Create artificial classification dataset

X, y = make_classification(

    n_samples=400,      # Total data points

    n_features=5,       # Total dimensions/features

    n_informative=4,    # Important features

    n_redundant=1,      # Redundant feature

    n_classes=3,        # Number of classes

    random_state=42
)

# Print dataset shape
print("Original Dataset Shape :", X.shape)

# Print target labels shape
print("Target Shape :", y.shape)
```

```python
# Plot first 2 original features

plt.scatter(

    X[:,0],     # Feature 1

    X[:,1],     # Feature 2

    c=y         # Color based on class labels
)

plt.xlabel("Feature 1")

plt.ylabel("Feature 2")

plt.title("Original Dataset")

plt.show()
```

```python
# LDA reduces dimensions while maximizing class separation

lda = LinearDiscriminantAnalysis(

    n_components=2     # Reduce dimensions to 2
)
```

```python
# fit_transform():
# 1. Learns class separation
# 2. Converts data into lower dimensions

X_lda = lda.fit_transform(X, y)

# Print reduced dataset shape
print("Reduced Dataset Shape :", X_lda.shape)
```

```python
    # Shows importance of each discriminant component

print("Explained Variance Ratio :")

print(lda.explained_variance_ratio_)
```

```python
# Scatter plot of reduced LDA data

plt.scatter(

    X_lda[:,0],      # LDA Component 1

    X_lda[:,1],      # LDA Component 2

    c=y
)

plt.xlabel("LDA Component 1")

plt.ylabel("LDA Component 2")

plt.title("LDA Reduced Dataset")

plt.show()
```

```python
# Store explained variance values

variance = lda.explained_variance_ratio_

# Labels for graph
components = ["LD1", "LD2"]

# Create bar graph
plt.bar(components, variance)

plt.xlabel("LDA Components")

plt.ylabel("Explained Variance")

plt.title("LDA Explained Variance")

plt.show()
```

```python
# scalings_
# Shows feature importance in discriminant directions

print("LDA Feature Scalings :")

print(lda.scalings_)
```
