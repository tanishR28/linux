```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

from sklearn.metrics import silhouette_score
```

```python
# Generate artificial clustered dataset

X, y = make_blobs(
    n_samples=400,
    centers=3,
    n_features=2,
    cluster_std=1.5,
    random_state=42
)

print("Dataset Shape :", X.shape)
```

```python
plt.scatter(X[:,0], X[:,1])

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Synthetic Dataset")

plt.show()
```

```python
# Gaussian Mixture Model (GMM)
# Implements Expectation Maximization Algorithm

em_model = GaussianMixture(
    n_components=3,     # Number of clusters
    random_state=42
)
```

```python
em_model.fit(X)

print("Model Training Completed")
```

```python
clusters = em_model.predict(X)

print("Cluster Labels :")

print(clusters)
```

```python
print("Cluster Means :")

print(em_model.means_)
```

```python
# Probability of each point belonging to clusters

probabilities = em_model.predict_proba(X)

print("Cluster Probabilities :")

print(probabilities[:5])
```

```python
score = silhouette_score(X, clusters)

print("Silhouette Score :", score)
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters)

# Plot cluster centers
plt.scatter(
    em_model.means_[:,0],
    em_model.means_[:,1],
    s=200,
    marker='X'
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Expectation Maximization Clustering")

plt.show()
```

```python
new_data = [[2, 3]]

prediction = em_model.predict(new_data)

print("Predicted Cluster :", prediction)
```
