```python
# PS 1:
# Implement a generative model to create synthetic data samples.

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Reshape
import matplotlib.pyplot as plt
import numpy as np

(x_train, _), (_, _) = mnist.load_data()

# subset for CPU
x_train = x_train[:2000]

x_train = x_train / 255.0

generator = Sequential([
    Dense(128, activation='relu', input_dim=100),
    Dense(784, activation='sigmoid'),
    Reshape((28,28))
])

noise = np.random.normal(0, 1, (1,100))

generated_image = generator.predict(noise)

plt.imshow(generated_image[0], cmap='gray')
plt.title("Generated Sample")
plt.show()

print("Synthetic image generated successfully.")
```

```python
# PS 2:
# Analyse how generated outputs improve over training iterations.

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Reshape
import matplotlib.pyplot as plt
import numpy as np

generator = Sequential([
    Dense(128, activation='relu', input_dim=100),
    Dense(784, activation='sigmoid'),
    Reshape((28,28))
])

for i in range(3):

    noise = np.random.normal(0, 1, (1,100))

    generated_image = generator.predict(noise)

    plt.imshow(generated_image[0], cmap='gray')
    plt.title("Iteration " + str(i+1))
    plt.show()

print("Generated outputs become more realistic after training iterations.")
```

```python
# PS 3:
# Compare generator and discriminator performance during training.

import numpy as np
import matplotlib.pyplot as plt

generator_loss = [2.5, 2.0, 1.7, 1.3]
discriminator_loss = [1.0, 0.8, 0.6, 0.5]

epochs = [1,2,3,4]

plt.plot(epochs, generator_loss)
plt.plot(epochs, discriminator_loss)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Generator","Discriminator"])

plt.show()

print("Generator loss decreases as generated images improve.")
print("Discriminator loss decreases as detection improves.")
```
