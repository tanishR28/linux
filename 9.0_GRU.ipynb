```python
# PS 1:
# Develop a sequence model optimised for faster training.

import time
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)

# subset for CPU
x_train = x_train[:2000]
y_train = y_train[:2000]

x_test = x_test[:500]
y_test = y_test[:500]

x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)

model = Sequential([
    Embedding(5000, 32, input_length=100),
    GRU(32),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

start = time.time()

model.fit(x_train, y_train, epochs=2)

end = time.time()

loss, acc = model.evaluate(x_test, y_test)

print("Accuracy:", acc)
print("Training Time:", end - start, "seconds")
print("GRU provides faster training with good performance.")
```

```python
# PS 2:
# Compare performance and training time with another recurrent architecture.

import time
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)

x_train = x_train[:2000]
y_train = y_train[:2000]

x_test = x_test[:500]
y_test = y_test[:500]

x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)

# GRU Model
gru_model = Sequential([
    Embedding(5000, 32, input_length=100),
    GRU(32),
    Dense(1, activation='sigmoid')
])

gru_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

start = time.time()

gru_model.fit(x_train, y_train, epochs=2, verbose=0)

gru_time = time.time() - start

gru_loss, gru_acc = gru_model.evaluate(x_test, y_test)

# LSTM Model
lstm_model = Sequential([
    Embedding(5000, 32, input_length=100),
    LSTM(32),
    Dense(1, activation='sigmoid')
])

lstm_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

start = time.time()

lstm_model.fit(x_train, y_train, epochs=2, verbose=0)

lstm_time = time.time() - start

lstm_loss, lstm_acc = lstm_model.evaluate(x_test, y_test)

print("GRU Accuracy:", gru_acc)
print("GRU Training Time:", gru_time)

print("LSTM Accuracy:", lstm_acc)
print("LSTM Training Time:", lstm_time)

print("GRU trains faster while LSTM may achieve slightly better accuracy.")
```

```python
# PS 3:
# Analyse efficiency and accuracy trade-offs.

import time
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)

x_train = x_train[:2000]
y_train = y_train[:2000]

x_test = x_test[:500]
y_test = y_test[:500]

x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)

model = Sequential([
    Embedding(5000, 32, input_length=100),
    GRU(16),   # smaller GRU for efficiency
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

start = time.time()

model.fit(x_train, y_train, epochs=2)

end = time.time()

loss, acc = model.evaluate(x_test, y_test)

print("Accuracy:", acc)
print("Training Time:", end - start)

print("Smaller GRU improves efficiency but may slightly reduce accuracy.")
```

```python
# PS 4:
# Evaluate the suitability of the model for real-time applications.

import time
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)

x_train = x_train[:1000]
y_train = y_train[:1000]

x_test = x_test[:100]
y_test = y_test[:100]

x_train = pad_sequences(x_train, maxlen=80)
x_test = pad_sequences(x_test, maxlen=80)

model = Sequential([
    Embedding(5000, 16, input_length=80),
    GRU(16),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train, y_train, epochs=2)

start = time.time()

pred = model.predict(x_test[:1])

end = time.time()

print("Prediction:", pred)
print("Prediction Time:", end - start, "seconds")

print("GRU is suitable for real-time applications because of faster inference.")
```
