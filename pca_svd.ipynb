```python
# NumPy -> Numerical operations
import numpy as np

# Matplotlib -> Data visualization / graphs
import matplotlib.pyplot as plt

# make_classification -> Generates synthetic dataset
from sklearn.datasets import make_classification

# PCA -> Dimensionality Reduction Technique
from sklearn.decomposition import PCA
```

```python
# Create artificial dataset having 5 features

X, y = make_classification(

    n_samples=400,      # Total number of data points

    n_features=5,       # Total number of features/dimensions

    n_informative=4,    # Useful features carrying information

    n_redundant=1,      # Redundant/correlated feature

    random_state=42     # Same output every run
)

# Print dataset shape
print("Original Dataset Shape :", X.shape)
```

```python
# Scatter plot using first 2 features

plt.scatter(

    X[:,0],     # Feature 1 values

    X[:,1],     # Feature 2 values

    c=y         # Color based on class labels
)

# Label X-axis
plt.xlabel("Feature 1")

# Label Y-axis
plt.ylabel("Feature 2")

# Graph title
plt.title("Original Dataset")

# Display graph
plt.show()
```

```python
# PCA reduces dimensions while preserving important information

pca = PCA(

    n_components=2     # Reduce data from 5 dimensions -> 2 dimensions
)
```

```python
# fit_transform():
# 1. Learns principal components
# 2. Converts original data into reduced dimensions

X_pca = pca.fit_transform(X)

# Print reduced dataset shape
print("Reduced Dataset Shape :", X_pca.shape)
```

```python
# explained_variance_ratio_
# Shows how much information each principal component preserves

print("Explained Variance Ratio :")

print(pca.explained_variance_ratio_)
```

```python
# Plot reduced 2D PCA data

plt.scatter(

    X_pca[:,0],    # Principal Component 1

    X_pca[:,1],    # Principal Component 2

    c=y
)

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA Reduced Dataset")

plt.show()
```

```python
# Store explained variance values

variance = pca.explained_variance_ratio_

# Labels for graph
components = ["PC1", "PC2"]

# Create bar graph
plt.bar(components, variance)

plt.xlabel("Principal Components")

plt.ylabel("Explained Variance")

plt.title("PCA Explained Variance")

plt.show()
```

```python
# TruncatedSVD -> Performs dimensionality reduction using SVD

from sklearn.decomposition import TruncatedSVD
```

```python
# Generate artificial dataset again

X, y = make_classification(

    n_samples=400,

    n_features=5,

    n_informative=4,

    n_redundant=1,

    random_state=42
)

print("Original Dataset Shape :", X.shape)
```

```python
# Create SVD model

svd = TruncatedSVD(

    n_components=2     # Reduce dimensions to 2
)
```

```python
# fit_transform():
# Learns important components and reduces dimensions

X_svd = svd.fit_transform(X)

print("Reduced Dataset Shape :", X_svd.shape)
```

```python
# Print variance captured by each SVD component

print("Explained Variance Ratio :")

print(svd.explained_variance_ratio_)
```

```python
# Scatter plot for reduced SVD data

plt.scatter(

    X_svd[:,0],

    X_svd[:,1],

    c=y
)

plt.xlabel("Component 1")

plt.ylabel("Component 2")

plt.title("SVD Reduced Dataset")

plt.show()
```

```python
# Store explained variance

variance = svd.explained_variance_ratio_

# Labels for graph
components = ["Component 1", "Component 2"]

# Create bar graph
plt.bar(components, variance)

plt.xlabel("Components")

plt.ylabel("Explained Variance")

plt.title("SVD Explained Variance")

plt.show()
```
